```python
#1
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def detect_and_remove_loop(head):
    slow = head
    fast = head

    # Move the two pointers until they meet
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break

    # If there is no loop, return
    if slow != fast:
        return

    # Move one pointer to the head of the linked list
    # and keep the other pointer at the meeting point
    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    # Unlink the last node from the loop
    fast.next = None

# Create a linked list with a loop for testing
# (Assuming the loop starts at position 3)

head = Node(1)
second = Node(2)
third = Node(3)
fourth = Node(4)
fifth = Node(5)

head.next = second
second.next = third
third.next = fourth
fourth.next = fifth
fifth.next = third  # Loop created

# Call the function to remove the loop
detect_and_remove_loop(head)

# Print the linked list to verify the loop has been removed
current = head
while current:
    print(current.data, end=" ")
    current = current.next
```

```python
#2
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def add_one(head):
    # Reverse the linked list
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    head = prev

    # Traverse the reversed linked list and add 1
    current = head
    carry = 1
    while current:
        current.data += carry
        if current.data == 10:
            current.data = 0
        else:
            carry = 0
            break
        current = current.next

    # If all nodes became 9, insert a new node with value 1
    if carry == 1:
        new_node = Node(1)
        new_node.next = head
        head = new_node

    # Reverse the linked list again
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    head = prev

    return head

# Create a linked list representing the number 123

head = Node(1)
second = Node(2)
third = Node(3)

head.next = second
second.next = third

# Call the function to add 1 to the linked list
new_head = add_one(head)

# Print the updated linked list
current = new_head
while current:
    print(current.data, end=" ")
    current = current.next
```

```python
#3
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.bottom = None

def merge_lists(list1, list2):
    # Base cases
    if not list1:
        return list2
    if not list2:
        return list1

    # Compare the data of the nodes
    if list1.data < list2.data:
        result = list1
        result.bottom = merge_lists(list1.bottom, list2)
    else:
        result = list2
        result.bottom = merge_lists(list1, list2.bottom)

    return result

def flatten(head):
    # Base case: an empty list or a list with a single node
    if not head or not head.next:
        return head

    # Recursively flatten the remaining list
    head.next = flatten(head.next)

    # Merge the current list with the flattened list
    head = merge_lists(head, head.next)

    return head

# Create a sample linked list with sub-linked lists

# Sub-list 1: 5 -> 7 -> 8
head = Node(5)
head.bottom = Node(7)
head.bottom.bottom = Node(8)

# Sub-list 2: 10 -> 12
head.next = Node(10)
head.next.bottom = Node(12)

# Sub-list 3: 15 -> 16 -> 20
head.next.next = Node(15)
head.next.next.bottom = Node(16)
head.next.next.bottom.bottom = Node(20)

# Call the function to flatten the linked list
flattened_head = flatten(head)

# Print the flattened linked list
current = flattened_head
while current:
    print(current.data, end=" ")
    current = current.bottom
```

```python
#4
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.random = None

def copy_random_list(head):
    if not head:
        return None

    # Step 1: Create a copy of each node and insert it after its original node
    current = head
    while current:
        new_node = Node(current.data)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next

    # Step 2: Update the random pointers of the copied nodes
    current = head
    while current:
        if current.random:
            current.next.random = current.random.next
        current = current.next.next

    # Step 3: Separate the original and copied lists
    current = head
    new_head = head.next
    copied_current = new_head
    while current:
        current.next = current.next.next
        if copied_current.next:
            copied_current.next = copied_current.next.next
        current = current.next
        copied_current = copied_current.next

    return new_head

# Create a sample linked list with random pointers

# Original List: 1 -> 2 -> 3 -> 4 -> 5
head = Node(1)
second = Node(2)
third = Node(3)
fourth = Node(4)
fifth = Node(5)

head.next = second
second.next = third
third.next = fourth
fourth.next = fifth

# Set random pointers
head.random = third  # 1 -> 3
second.random = fifth  # 2 -> 5
third.random = second  # 3 -> 2
fourth.random = third  # 4 -> 3
fifth.random = fourth  # 5 -> 4

# Call the function to create a deep copy of the linked list
new_head = copy_random_list(head)

# Print the copied linked list
current = new_head
while current:
    print("Node:", current.data)
    print("Next Pointer:", current.next.data if current.next else None)
    print("Random Pointer:", current.random.data if current.random else None)
    print()
    current = current.next
```

```python
#5
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def odd_even_list(head):
    if not head or not head.next:
        return head

    odd_head = odd = head
    even_head = even = head.next

    while even and even.next:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next

    odd.next = even_head
    return odd_head

# Example usage:
# Create a sample linked list: 1 -> 2 -> 3 -> 4 -> 5 -> None
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

# Call the function to group nodes with odd and even indices
result = odd_even_list(head)

# Print the reordered list
current = result
while current:
    print(current.val, end=" ")
    current = current.next
```

```python
#6
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def left_shift_list(head, k):
    if not head or not head.next or k == 0:
        return head

    length = 0
    current = head
    while current.next:
        length += 1
        current = current.next

    k = k % (length + 1)  # Adjust k if it exceeds the length of the list

    if k == 0:
        return head

    current = head
    for _ in range(k):
        current = current.next

    new_head = current
    while current.next:
        current = current.next

    current.next = head
    current = new_head
    for _ in range(k - 1):
        current = current.next

    current.next = None

    return new_head

# Example usage:
# Create a sample linked list: 1 -> 2 -> 3 -> 4 -> 5 -> None
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

k = 2

# Call the function to left-shift the linked list by k nodes
result = left_shift_list(head, k)

# Print the shifted list
current = result
while current:
    print(current.val, end=" ")
    current = current.next
```

```python
#7
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def next_greater_node(head):
    # Convert linked list to list
    nodes = []
    while head:
        nodes.append(head.val)
        head = head.next

    n = len(nodes)
    result = [0] * n
    stack = []

    for i in range(n - 1, -1, -1):
        while stack and nodes[i] >= stack[-1]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(nodes[i])

    return result

# Example usage:
# Create a sample linked list: 2 -> 1 -> 5 -> None
head = ListNode(2)
head.next = ListNode(1)
head.next.next = ListNode(5)

# Call the function to find the next greater node for each node
result = next_greater_node(head)

# Print the result
print(result)
```

```python
#8
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def remove_zero_sum_sublists(head):
    dummy = ListNode(0)
    dummy.next = head
    cumulative_sum = 0
    sum_map = {}

    current = dummy
    while current:
        cumulative_sum += current.val

        if cumulative_sum in sum_map:
            prev = sum_map[cumulative_sum].next
            remove_sum = cumulative_sum
            while prev != current:
                remove_sum += prev.val
                sum_map.pop(remove_sum)
                prev = prev.next

            sum_map[cumulative_sum].next = current.next
        else:
            sum_map[cumulative_sum] = current

        current = current.next

    return dummy.next

# Example usage:
# Create a sample linked list: 1 -> 2 -> -3 -> 3 -> 1 -> None
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(-3)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(1)

# Call the function to remove consecutive sequences of nodes with a sum of 0
result = remove_zero_sum_sublists(head)

# Print the final linked list
current = result
while current:
    print(current.val, end=" ")
    current = current.next
```
