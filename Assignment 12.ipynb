```python
#1
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def deleteMiddleNode(head):
    if head is None or head.next is None:
        return None

    slow = head
    fast = head
    prev = None

    while fast is not None and fast.next is not None:
        fast = fast.next.next
        prev = slow
        slow = slow.next

    prev.next = slow.next

    return head

# Test the function
# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Original Linked List:")
current = head
while current is not None:
    print(current.val, end=" ")
    current = current.next

head = deleteMiddleNode(head)

print("\nModified Linked List:")
current = head
while current is not None:
    print(current.val, end=" ")
    current = current.next
```

```python
#2
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def hasCycle(head):
    if head is None or head.next is None:
        return False

    slow = head
    fast = head.next

    while slow != fast:
        if fast is None or fast.next is None:
            return False

        slow = slow.next
        fast = fast.next.next

    return True

# Test the function
# Create a linked list with a loop: 1 -> 2 -> 3 -> 4 -> 5 -> 2 (loop)
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = head.next

has_loop = hasCycle(head)
if has_loop:
    print("The linked list has a loop.")
else:
    print("The linked list does not have a loop.")
```

```python
#3
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def findNthFromEnd(head, n):
    if head is None:
        return None

    # Initialize two pointers: slow and fast
    slow = head
    fast = head

    # Move the fast pointer n nodes ahead
    for _ in range(n):
        if fast is None:
            return None
        fast = fast.next

    # Move both pointers until the fast pointer reaches the end
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next

    # The slow pointer is now pointing to the Nth node from the end
    return slow.val

# Test the function
# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

n = 2
nth_node = findNthFromEnd(head, n)
if nth_node is None:
    print("Invalid value of N.")
else:
    print(f"The {n}th node from the end is: {nth_node}")
```

```python
#5
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def detectAndRemoveLoop(head):
    if head is None or head.next is None:
        return

    # Detect the loop using Floyd's cycle-finding algorithm
    slow = head
    fast = head
    has_loop = False
    while fast.next is not None and fast.next.next is not None:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            has_loop = True
            break

    # If there is no loop, return
    if not has_loop:
        return

    # Find the start of the loop
    slow = head
    while slow != fast:
        slow = slow.next
        fast = fast.next

    # Find the last node of the loop
    last_node = fast
    while last_node.next != slow:
        last_node = last_node.next

    # Remove the loop by setting the next pointer of the last node to None
    last_node.next = None

# Test the function
# Create a linked list with a loop: 1 -> 2 -> 3 -> 4 -> 5 -> 3 (loop)
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = head.next.next  # Creating a loop

# Detect and remove the loop
detectAndRemoveLoop(head)

# Print the modified linked list
current = head
while current is not None:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```

```python
#6
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def deleteNodes(head, M, N):
    # Edge case: Empty linked list
    if head is None:
        return None

    # Start with the head of the linked list
    current = head

    # Traverse the linked list
    while current is not None:
        # Retain M nodes
        for _ in range(M - 1):
            if current is None:
                break
            current = current.next

        # Check if M nodes were retained
        if current is None:
            break

        # Delete N nodes
        next_node = current.next
        for _ in range(N):
            if next_node is None:
                break
            next_node = next_node.next
        current.next = next_node

        # Move to the next node
        current = next_node

    return head

# Create a linked list: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8 -> 9 -> None
head = ListNode(1)
current = head
for i in range(2, 10):
    current.next = ListNode(i)
    current = current.next

# Call deleteNodes with M=2 and N=3
M = 2
N = 3
result = deleteNodes(head, M, N)

# Print the modified linked list
current = result
while current is not None:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```

```python
#7
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeLists(head1, head2):
    # Edge case: Empty first list
    if head1 is None:
        return head2

    current1 = head1
    current2 = head2

    while current1 is not None and current2 is not None:
        # Save the next pointers
        next1 = current1.next
        next2 = current2.next

        # Insert the current node of the second list into the first list
        current1.next = current2
        current2.next = next1

        # Move to the next nodes
        current1 = next1
        current2 = next2

    return head1

# Example usage
# Create the first linked list: 5 -> 7 -> 17 -> 13 -> 11 -> None
head1 = ListNode(5)
head1.next = ListNode(7)
head1.next.next = ListNode(17)
head1.next.next.next = ListNode(13)
head1.next.next.next.next = ListNode(11)

# Create the second linked list: 12 -> 10 -> 2 -> 4 -> 6 -> None
head2 = ListNode(12)
head2.next = ListNode(10)
head2.next.next = ListNode(2)
head2.next.next.next = ListNode(4)
head2.next.next.next.next = ListNode(6)

# Merge the lists
merged_head = mergeLists(head1, head2)

# Print the merged list
current = merged_head
while current is not None:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```

```python
#8
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def isCircular(head):
    # Edge case: Empty list
    if head is None:
        return False

    slow = head
    fast = head.next

    while fast is not None and fast.next is not None:
        if slow == fast:
            return True

        slow = slow.next
        fast = fast.next.next

    return False

# Example 
head = ListNode(10)
head.next = ListNode(12)
head.next.next = ListNode(14)
head.next.next.next = ListNode(16)
head.next.next.next.next = head.next

# Check if the linked list is circular
result = isCircular(head)

if result:
    print("The linked list is circular.")
else:
    print("The linked list is not circular.")
```
