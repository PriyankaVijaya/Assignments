```python
#1
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def createNewLinkedList(head1, head2):
    if not head1:
        return head2
    if not head2:
        return head1

    resultHead = None
    resultTail = None

    while head1 and head2:
        if head1.data >= head2.data:
            newNode = Node(head1.data)
            head1 = head1.next
        else:
            newNode = Node(head2.data)
            head2 = head2.next

        if not resultHead:
            resultHead = newNode
            resultTail = newNode
        else:
            resultTail.next = newNode
            resultTail = newNode

    if head1:
        resultTail.next = head1
    if head2:
        resultTail.next = head2

    return resultHead

# Example usage
# Create linked list 1: 1 -> 4 -> 5 -> 8
head1 = Node(1)
head1.next = Node(4)
head1.next.next = Node(5)
head1.next.next.next = Node(8)

# Create linked list 2: 2 -> 3 -> 6 -> 7
head2 = Node(2)
head2.next = Node(3)
head2.next.next = Node(6)
head2.next.next.next = Node(7)

# Create a new linked list using the greater nodes
resultHead = createNewLinkedList(head1, head2)

# Print the new linked list
current = resultHead
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#2
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def removeDuplicates(head):
    if not head or not head.next:
        return head

    current = head
    while current.next:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next

    return head

# Example usage
# Create linked list: 11 -> 11 -> 11 -> 21 -> 43 -> 43 -> 60
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)

# Remove duplicate nodes
newHead = removeDuplicates(head)

# Print the modified linked list
current = newHead
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#3
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def reverseKNodes(head, k):
    if not head or k == 1:
        return head

    dummy = Node(0)
    dummy.next = head
    prev = dummy

    while True:
        count = 0
        current = prev.next
        next = None

        # Find the kth node or the end of the linked list
        while count < k and current:
            next = current.next
            count += 1

        if count == k:
            # Reverse the k nodes
            prev.next = reverseLinkedList(prev.next, k)

            # Update pointers for the reversed group
            for _ in range(k):
                prev = prev.next
        else:
            break

    return dummy.next

def reverseLinkedList(head, k):
    prev = None
    current = head

    while k > 0:
        next = current.next
        current.next = prev
        prev = current
        current = next
        k -= 1

    head.next = current

    return prev

# Example usage
# Create linked list: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8 -> 9
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)

# Reverse every 3 nodes
k = 3
newHead = reverseKNodes(head, k)

# Print the modified linked list
current = newHead
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#4
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def reverseAlternateKNodes(head, k):
    if not head or k <= 1:
        return head

    dummy = Node(0)
    dummy.next = head
    prev = dummy
    current = head
    is_reverse = True

    while current:
        count = 0
        group_start = current
        prev_group_end = prev

        # Traverse k nodes or until the end of the linked list
        while current and count < k:
            next = current.next

            if is_reverse:
                # Reverse the current node
                current.next = prev
                prev = current
            else:
                prev = current

            current = next
            count += 1

        if is_reverse:
            # Update pointers for the reversed group
            group_start.next = current
            prev_group_end.next = prev

            # Move k steps forward
            for _ in range(k):
                prev = prev.next

        is_reverse = not is_reverse

    return dummy.next

# Example usage
# Create linked list: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8 -> 9
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)

# Reverse every alternate 3 nodes
k = 3
newHead = reverseAlternateKNodes(head, k)

# Print the modified linked list
current = newHead
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#5
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def deleteLastOccurrence(head, key):
    if not head:
        return None

    prev = None
    to_delete = None
    current = head

    # Traverse the linked list to find the last occurrence of the key
    while current:
        if current.data == key:
            to_delete = current
        current = current.next

    if not to_delete:
        # Key not found, return the original head
        return head

    current = head

    # Traverse the linked list again to update pointers
    while current:
        if current == to_delete:
            if prev:
                prev.next = current.next
            else:
                # Node to be deleted is the head
                head = current.next
            break
        prev = current
        current = current.next

    return head

# Example usage
# Create linked list: 1 -> 2 -> 3 -> 2 -> 4 -> 2 -> 5
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(2)
head.next.next.next.next = Node(4)
head.next.next.next.next.next = Node(2)
head.next.next.next.next.next.next = Node(5)

# Delete last occurrence of key 2
key = 2
newHead = deleteLastOccurrence(head, key)

# Print the modified linked list
current = newHead
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#6
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def mergeLists(head1, head2):
    # Check for empty lists
    if not head1:
        return head2
    if not head2:
        return head1

    # Set the head of the merged list
    if head1.data <= head2.data:
        mergedHead = head1
        mergedTail = head1
        ptr1 = head1.next
        ptr2 = head2
    else:
        mergedHead = head2
        mergedTail = head2
        ptr1 = head1
        ptr2 = head2.next

    # Merge the lists
    while ptr1 and ptr2:
        if ptr1.data <= ptr2.data:
            mergedTail.next = ptr1
            ptr1 = ptr1.next
        else:
            mergedTail.next = ptr2
            ptr2 = ptr2.next

        mergedTail = mergedTail.next

    # Append remaining nodes from list 1 or list 2
    if ptr1:
        mergedTail.next = ptr1
    elif ptr2:
        mergedTail.next = ptr2

    return mergedHead

# Example usage
# Create linked list 1: 1 -> 3 -> 5 -> 7 -> 9
head1 = Node(1)
head1.next = Node(3)
head1.next.next = Node(5)
head1.next.next.next = Node(7)
head1.next.next.next.next = Node(9)

# Create linked list 2: 2 -> 4 -> 6 -> 8
head2 = Node(2)
head2.next = Node(4)
head2.next.next = Node(6)
head2.next.next.next = Node(8)

# Merge the two linked lists
mergedHead = mergeLists(head1, head2)

# Print the merged list
current = mergedHead
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")
```

```python
#7
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

def reverseDoublyLinkedList(head):
    current = head
    while current:
        current.prev, current.next = current.next, current.prev
        if not current.prev:
            break
        current = current.prev
    return current

# Example usage
# Create doubly linked list: 1 <-> 2 <-> 3 <-> 4 <-> 5
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next
head.next.next.next.next = Node(5)
head.next.next.next.next.prev = head.next.next.next

# Reverse the doubly linked list
reversedHead = reverseDoublyLinkedList(head)

# Print the reversed doubly linked list
current = reversedHead
while current:
    print(current.data, end=" <-> ")
    current = current.next
print("None")

```

```python
#8
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

def deleteNode(head, position):
    if not head:
        return head
    
    # Case 1: If the node to be deleted is the head node
    if position == 1:
        if head.next:
            head.next.prev = None
        head = head.next
        return head
    
    current = head
    count = 1
    
    # Traverse to the position to be deleted
    while current and count < position:
        current = current.next
        count += 1
    
    # Case 2: If the node to be deleted is the last node
    if not current.next:
        current.prev.next = None
        return head
    
    # Case 3: If the node to be deleted is in between
    current.prev.next = current.next
    current.next.prev = current.prev
    
    return head

# Example usage
# Create doubly linked list: 1 <-> 2 <-> 3 <-> 4 <-> 5
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next
head.next.next.next.next = Node(5)
head.next.next.next.next.prev = head.next.next.next

# Delete node at position 3
position = 3
head = deleteNode(head, position)

# Print the updated doubly linked list
current = head
while current:
    print(current.data, end=" <-> ")
    current = current.next
print("None")
```
